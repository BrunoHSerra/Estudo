***Estudo Machine Learnig ***

**Definição:**

##
 Machine learning é o campo que estuda a capacidade do computador/máquina aprender sem a necessidade de ser diretamente programada.

 Uma outra boa definição é que podemos dizer que uma máquina pode aprender uma experiência E, dado uma tarefa T e um medidor de performance P, desde que sua performance em T, dado um medidor P, aumente a experiência E.


Um exemplo é um classificador de spam. Gostaríamos que o classificador marcasse os possíveis spams na entrada dos e-mails. Utilizamos para essa finalidade um conjunto de e-mails identificado como spam (geralmente feito por usuários) e um conjunto de e-mails ham (e-mails não considerados spam).
O conjunto utilizado para essa tarefa chamamos de trainnig set, ou conjunto de treinamento. Cada instância de treinamento é chamado de trainnig intance ou sample.
Para este exemplo a tarefa é marcar os spams, a experiência é o conjunto de dados e o parâmetro de performance é definido para cada caso. Neste tipo de problema geralmente é a acurácia, mas pode ser uma proporção ( ratio) por exemplo.

**Porque usar ML?**

Machine learnig facilita a descoberta/visualização de padrões, possibilitando que tarefas que demanda muita codificação sejam feita através da automação desses processos.
Logo temos um bom desempenho para as seguintes atividades:
- Problemas para os quais as soluções atuais demandam de um atividade que envolve muita codificação para sua definição.
- Problemas complexos para os quais não temos ainda uma boa abordagem para solução.
- Sistemas flutuantes, onde temos uma diversidade de tipo de dados.
- Obtenção de insights sobre problemas complexos e/ou sobre uma grande quantidade de dados.



**Tipos de sistemas de Machine Learning**

- Os quais são ou não treinados por humanos (supervisionado, não supervisionado, semi-supervisionado ou aprendizado por reforço).
- Se eles podem ou não aprender em tempo real (online ou batch learning).
- Os que trabalham comparando novos dados com dados já conhecidos ou os que detectam padrões nos dados de treinamento e geram um modelo preditivo ( instace-base versus model-based learning).

Esses pontos não são necessariamente separados, podendo ocorrer a mistura entre metodos.

**Aprendizado Semi-Supervisionado**

Este tipo de aprendizado ocorre quando temos um conjunto de dados onde alguns deles são rotulados e os outros não. Neste tipo de tarefa o algoritimo usa clustering para fazer agrupamentos e com esse dados agrupados ele usa os dados rotulados para classificar os demais. Exemplo disso é o google fotos, se você postar fotos com outras pessoas o algoritmo começa a juntar as imagens dessas pessoas em clusters, no momento em que essa pessoa for rotulada em alguma foto ele vai conseguir nomaear a pessoa nas demais fotos.

# **Batch Learning e Online Learning**

##**Batch Learning**

Neste método o sistema é incapaz de aprender incrementalmente, é necessário aprender de uma base de dados existente, demandando assim muito tempo e poder computacional, por isso essencialmente é feito offline.
Primeiramente o modelo é treinado e em seguida lançado para produção, colocando em prática assim o que foi aprendido no treinamento.
Se quisermos que o modelo volte a aprender temos que parar a execução e treinar o modelo offline do zero novamente inserindo novos dados, desse modo alteramos o programa anterior, lançando uma nova versão. Felizmente esse processo treinamento, avaliação e lançamento pode ser automatizado e executado de maneira simples. Porém, o processo de treinamento do zero com cada vez mais dados é muito caro (computacionalmente) e demorado, logo deve ser feito a cada 24 horas ou semanalmente.

##**Online Learning**

Neste tipo de treinamento é introduzida uma etapa, pós treinamento inicial, onde são inseridos sets ou pacotes de sets, de forma sequencial, diminuindo assim a necessidade computaticional para efetuar o treinamento.

image.png

Este tipo de estratégia benéfica para capacidade computacional reduzida, pode-se ir incrementando aos poucos o dataset no treinamento a medida que o sistema já está rodando. Desse modo é reduzido o consumo de memória no treinamento inicial.

image.png

Um ponto crucial neste método é a taxa de apredizagem. Se a taxa for alta o modelo irá aprender rápido mas também vai esquecer rápido o que foi aprendido antes. Já se tivermos uma taxa baixa vamos ter uma inercia inicial alta, mas em contrapartida ficamos com os dados armazenados por mais tempo.
O grande desafio neste modelo é a implementação dos dados reais corretos, uma vez que o upload de dados errados ou inviesados vai perdurar por muito tempo no modelo levando a um declínio da assertivadade, bem como, a problemas que podem ser identificados em tempo real pelo usuáriuo final.
